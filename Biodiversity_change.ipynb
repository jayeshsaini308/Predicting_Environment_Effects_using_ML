### Extracting Climatic variables from the TerraClimate sample GeoTiff image

```python
# Extracts satellite variables from a GeoTIFF based on coordinates from a csv file and returns them in a DataFrame.

def map_satellite_data(tiff_path, csv_path):
    
    df = pd.read_csv(csv_path)

    # Open the GeoTIFF file and load data into xarray DataArrays
    with rasterio.open(tiff_path) as dataset:
        # Read raster variables into numpy arrays
        srad_data = dataset.read(1)  # variable 1 (e.g., srad)
        vap_data = dataset.read(2)   # variable 2 (e.g., vap)

        # Generate coordinate grids (lon/lat) from raster bounds and resolution
        lon = np.linspace(dataset.bounds.left, dataset.bounds.right, dataset.width)
        lat = np.linspace(dataset.bounds.top, dataset.bounds.bottom, dataset.height)

        # Create xarray DataArrays for each variable
        srad_da = xr.DataArray(
            srad_data, coords=[("lat", lat), ("lon", lon)], dims=["lat", "lon"]
        )
        vap_da = xr.DataArray(
            vap_data, coords=[("lat", lat), ("lon", lon)], dims=["lat", "lon"]
        )

    # Initialize lists to store extracted values
    srad_values = []
    vap_values = []

    # Extract values for each row in the DataFrame
    for _, row in tqdm(df.iterrows(), total=df.shape[0], desc="Extracting raster values"):
        target_lat, target_lon = row["Latitude"], row["Longitude"]

        try:
            # Extract the nearest values from the DataArrays
            srad_value = srad_da.sel(lat=target_lat, lon=target_lon, method="nearest").values
            vap_value = vap_da.sel(lat=target_lat, lon=target_lon, method="nearest").values
        except KeyError:
            # Handle cases where coordinates are outside raster bounds
            srad_value, vap_value = np.nan, np.nan

        srad_values.append(srad_value)
        vap_values.append(vap_value)
        
    # Create a DataFrame to store the variable values
    df1 = pd.DataFrame()
    df1["srad"] = srad_values
    df1["vap"] = vap_values

    return df1
```

```python
# Combine two datasets vertically (along columns) using pandas concat function.
def combine_two_datasets(dataset1,dataset2):
    '''
    Returns a  vertically concatenated dataset.
    Attributes:
    dataset1 - Dataset 1 to be combined 
    dataset2 - Dataset 2 to be combined
    '''
    
    data = pd.concat([dataset1,dataset2], axis=1)
    return data
```

### Building and Training the Machine Learning Pipeline

In this section, we define a machine learning pipeline that includes both preprocessing and model training steps. First, we specify the numerical features (`srad` and `vap`) and set up a preprocessing pipeline (`num_transformer`). This pipeline applies a `SimpleImputer` to handle any missing values by replacing them with the median, followed by a `MinMaxScaler` to normalize the data. Then, we combine this preprocessing step using `ColumnTransformer`. Next, we define a Logistic Regression model, which will be trained on the processed data. The entire workflow, including preprocessing and model fitting, is encapsulated in a `Pipeline`. We then split the dataset into training and test sets, using 70% of the data for training and 30% for testing. Finally, we train the pipeline using the training data. This ensures that preprocessing and model fitting occur in a streamlined and consistent manner.

```python
features = ["srad", "vap"]

# Define preprocessing steps
transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),  # Fill missing values with mean
    ("scaler", MinMaxScaler())                 # Scale numerical features
])


# Combine transformers using ColumnTransformer
preprocessor = ColumnTransformer([
    ("num", transformer, features)
])

# Define the Logistic Regression model
model = LogisticRegression(max_iter=1000)

# Create the full pipeline
pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("classifier", model)
])

# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=42)

# Train the pipeline
pipeline.fit(X_train, y_train)
```

```python

terraclimate_file = "terraclimate_median_all_variables.tif"

# Read the GeoTIFF file using rasterio
with rasterio.open(terraclimate_file) as src:
    # Read all bands (variables) from the file
    terraclimate_data = src.read()
    # Get the metadata including the transform for geolocation
    transform = src.transform
    # Get the variable names (assuming they're stored in band descriptions)
    variable_names = [src.descriptions[i] for i in range(src.count)]

# Function to extract values from all bands at given coordinates
def extract_values_all_vars(lon, lat):
    # Transform coordinates to pixel indices
    py, px = rasterio.transform.rowcol(transform, lon, lat)
    
    # Initializing a dictionary to store all variable values
    values = {}
    
    # Extracting values for each band/variable
    for i, var_name in enumerate(variable_names):
        # Check if the point is within the raster bounds
        if 0 <= py < terraclimate_data.shape[1] and 0 <= px < terraclimate_data.shape[2]:
            values[var_name] = terraclimate_data[i, py, px]
        else:
            values[var_name] = np.nan  # Handle points outside the raster
    
    return values

# Extract all climate variables for each point in the ground data
climate_data = []
for idx, row in tqdm(ground_df.iterrows(), total=len(ground_df)):
    lon, lat = row['Longitude'], row['Latitude']
    values = extract_values_all_vars(lon, lat)
    climate_data.append(values)

# Convert to DataFrame
climate_df = pd.DataFrame(climate_data)

# Combining with the original ground data
full_df = pd.concat([ground_df, climate_df], axis=1)

# Drop rows with missing values
full_df.dropna(inplace=True)

# Preparing features and target
X = full_df.drop(['Latitude', 'Longitude', 'Occurrence Status'], axis=1)
y = full_df['Occurrence Status']

# Spliting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# preprocessing pipeline
numeric_features = X.columns.tolist()
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)
    ])

# Creating and training the model pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42))
])

model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
disp = ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
disp.plot()
plt.show()

# Feature importance analysis (for logistic regression coefficients)
if hasattr(model.named_steps['classifier'], 'coef_'):
    # Geting feature names after preprocessing
    feature_names = numeric_features  # Since we're using all numeric features
    
    # Get coefficients
    coefficients = model.named_steps['classifier'].coef_[0]
    
    # Creating a DataFrame for visualization
    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': coefficients
    }).sort_values('Importance', ascending=False)
    
    print("\nFeature Importance:")
    print(importance_df)
    
    # Plot feature importance
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=importance_df)
    plt.title('Feature Importance (Logistic Regression Coefficients)')
    plt.tight_layout()
    plt.show()

# Saving model for future use
joblib.dump(model, 'frog_presence_model_all_vars.pkl')
```

```python
# Import enhanced libraries
import numpy as np
import pandas as pd
import xarray as xr
import rasterio
from tqdm import tqdm
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import GroupKFold, RandomizedSearchCV
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import make_pipeline
from xgboost import XGBClassifier
import joblib
import matplotlib.pyplot as plt

# Enhanced TerraClimate data loading (using median file and temporal features)
def load_terraclimate_data(file_path, ground_df):
    """Load and process TerraClimate data with temporal features"""
    with rasterio.open(file_path) as src:
        data = src.read()
        transform = src.transform
        variables = src.descriptions

    # Creating multi-temporal features
    climate_records = []
    for idx, row in tqdm(ground_df.iterrows(), total=len(ground_df)):
        py, px = rasterio.transform.rowcol(transform, row['Longitude'], row['Latitude'])
        if 0 <= py < data.shape[1] and 0 <= px < data.shape[2]:
            record = {
                'lat': row['Latitude'],
                'lon': row['Longitude'],
                'status': row['Occurrence Status']
            }
            # Extracting base variables
            for i, var in enumerate(variables):
                record[var] = data[i, py, px]
            
            # Adding temporal aggregations (12-month window)
            if idx >= 12:  # Ensure rolling window
                for var in ['ppt', 'tmax', 'pdsi']:
                    # Extracting the band index for the variable
                    try:
                        var_idx = variables.index(var)
                    except ValueError:
                        print(f"Variable '{var}' not found in variables list.")
                        continue  # Skip to the next variable

                    # Calculating rolling average and trend
                    record[f'{var}_12m_avg'] = np.nanmean([data[var_idx, py, px] for i in range(max(0, idx-12), idx)])
                    record[f'{var}_12m_trend'] = np.polyfit(range(max(0, idx-12), idx), [data[var_idx, py, px] for i in range(max(0, idx-12), idx)], 1)[0]
            else:
                # Seting temporal features to NaN for initial records
                for var in ['ppt', 'tmax', 'pdsi']:
                    record[f'{var}_12m_avg'] = np.nan
                    record[f'{var}_12m_trend'] = np.nan
            
            climate_records.append(record)
    
    return pd.DataFrame(climate_records).dropna()

# Load and prepare data
print("Loading and processing TerraClimate data...")
full_df = load_terraclimate_data("terraclimate_median_all_variables.tif", ground_df)

# Enhanced feature engineering
def add_climate_features(df):
    """Add derived climate features"""
    df['aridity_index'] = df['ppt'] / (df['pet'] + 1e-6)
    df['temp_range'] = df['tmax'] - df['tmin']
    df['summer_ppt'] = df.groupby(df.index // 12)['ppt'].transform(lambda x: x[5:8].sum())
    return df

full_df = add_climate_features(full_df)

# Preparing data splits with spatial blocking
full_df['spatial_block'] = (full_df['lat'].round(1).astype(str) + 
                           '_' + 
                           full_df['lon'].round(1).astype(str))

X = full_df.drop(['lat', 'lon', 'status', 'spatial_block'], axis=1)
y = full_df['status']
groups = full_df['spatial_block']

# Advanced preprocessing pipeline
numeric_features = X.columns.tolist()
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', RobustScaler())
        ]), numeric_features)
    ])

# Optimized XGBoost model with class balancing
model = make_pipeline(
    preprocessor,
    SMOTE(sampling_strategy='auto', random_state=42),
    XGBClassifier(
        objective='binary:logistic',
        tree_method='hist',
        enable_categorical=False,
        eval_metric='logloss',
        use_label_encoder=False,
        random_state=42
    )
)

# Hyperparameter search space
param_grid = {
    'xgbclassifier__n_estimators': [100, 200, 300, 400],
    'xgbclassifier__max_depth': [5, 7, 9, 11],
    'xgbclassifier__learning_rate': [0.005, 0.01, 0.05, 0.1],
    'xgbclassifier__subsample': [0.5, 0.6, 0.7, 0.8, 0.9],
    'xgbclassifier__colsample_bytree': [0.5, 0.6, 0.7, 0.8, 0.9],
    'xgbclassifier__gamma': [0, 0.1, 0.2, 0.3],
    'xgbclassifier__reg_alpha': [0, 0.1, 0.2, 0.3],  # L1 regularization
    'xgbclassifier__reg_lambda': [1, 1.1, 1.2, 1.3]   # L2 regularization
}

# Spatial cross-validation
cv = GroupKFold(n_splits=5)
search = RandomizedSearchCV(
    model, 
    param_grid, 
    n_iter=100,
    cv=cv,
    scoring='accuracy',
    n_jobs=-1,
    verbose=3,
    error_score='raise'  # Raise errors during hyperparameter tuning
)

print("Starting model training with spatial cross-validation...")
search.fit(X, y, groups=groups)

# Final evaluation
best_model = search.best_estimator_
print(f"Best Accuracy: {search.best_score_:.2f}")
print(f"Optimal Parameters: {search.best_params_}")

# optimized model
joblib.dump(best_model, 'optimized_frog_model.pkl')

# Feature importance visualization
feature_names = numeric_features
importances = best_model.named_steps['xgbclassifier'].feature_importances_

pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values('Importance', ascending=False).set_index('Feature').plot.barh()
plt.title('XGBoost Feature Importance')
plt.show()
```

```python
# Select Top 4 Features from Trained Model 
feature_names = numeric_features
importances = best_model.named_steps['xgbclassifier'].feature_importances_

feat_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values('Importance', ascending=False)

top_features = feat_df['Feature'].head(4).tolist()

# Creating new dataset using only top 4 features 
X_top = X[top_features]

# Outlier Removal using IQR Method 
def remove_outliers_iqr(df, feature_columns, threshold=1.5):
    clean_df = df.copy()
    for col in feature_columns:
        Q1 = clean_df[col].quantile(0.25)
        Q3 = clean_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - threshold * IQR
        upper = Q3 + threshold * IQR
        clean_df = clean_df[(clean_df[col] >= lower) & (clean_df[col] <= upper)]
    return clean_df

# Combining with target and groups for outlier filtering
combined_df = pd.concat([X_top.reset_index(drop=True), y.reset_index(drop=True), groups.reset_index(drop=True)], axis=1)
combined_df.columns = top_features + ['status', 'spatial_block']

clean_df = remove_outliers_iqr(combined_df, top_features)

# Final cleaned data
X_clean = clean_df[top_features]
y_clean = clean_df['status']
groups_clean = clean_df['spatial_block']

# Retraining Model on Cleaned Data with Top 4 Features 
# Updating preprocessing pipeline for only top 4 features
preprocessor_clean = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', RobustScaler())
        ]), top_features)
    ]
)

model_clean = make_pipeline(
    preprocessor_clean,
    SMOTE(sampling_strategy='auto', random_state=42),
    XGBClassifier(
        objective='binary:logistic',
        tree_method='hist',
        enable_categorical=False,
        eval_metric='logloss',
        use_label_encoder=False,
        random_state=42
    )
)

# Retraining using the same RandomizedSearchCV settings
search_clean = RandomizedSearchCV(
    model_clean,
    param_grid,
    n_iter=150,
    cv=GroupKFold(n_splits=5),
    scoring='accuracy',
    n_jobs=-1,
    verbose=3
)

print("\nRetraining model with top 4 features and cleaned data...")
search_clean.fit(X_clean, y_clean, groups=groups_clean)

# Final Evaluation
print(f"\nImproved Accuracy (Top 4 Features): {search_clean.best_score_:.4f}")
print(f"Optimal Params: {search_clean.best_params_}")

# Saving model
joblib.dump(search_clean.best_estimator_, 'frog_model_top4_cleaned.pkl')

# new feature importance
importances_top4 = search_clean.best_estimator_.named_steps['xgbclassifier'].feature_importances_
pd.DataFrame({
    'Feature': top_features,
    'Importance': importances_top4
}).sort_values('Importance', ascending=False).set_index('Feature').plot.barh()
plt.title('Feature Importance (Top 4 - Cleaned)')
plt.show()
```

```python
feature_names = numeric_features
importances = best_model.named_steps['xgbclassifier'].feature_importances_

feat_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values('Importance', ascending=False)

top_features = feat_df['Feature'].head(8).tolist()
print(f"Top 8 features: {top_features}")

# Cliping outliers in top features instead of removing
X_top = X[top_features].copy()
for col in top_features:
    lower = X_top[col].quantile(0.01)
    upper = X_top[col].quantile(0.99)
    X_top[col] = X_top[col].clip(lower, upper)

# preprocessing for selected features
preprocessor_top = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', RobustScaler())
        ]), top_features)
    ])

# Rebuilding model pipeline
model_top = make_pipeline(
    preprocessor_top,
    SMOTE(sampling_strategy='auto', random_state=42),
    XGBClassifier(
        objective='binary:logistic',
        tree_method='hist',
        enable_categorical=False,
        eval_metric='logloss',
        use_label_encoder=False,
        random_state=42
    )
)

# hyperparameter grid
search_top = RandomizedSearchCV(
    model_top, 
    param_distributions=param_grid, 
    n_iter=150,
    cv=cv,
    scoring='accuracy',
    n_jobs=-1,
    verbose=3,
    error_score='raise'
)

print("Retraining model with top 8 features and clipped outliers...")
search_top.fit(X_top, y, groups=groups)

# Final evaluation
best_model_top = search_top.best_estimator_
print(f"Improved Accuracy (Top 8 Features): {search_top.best_score_:.4f}")
print(f"Optimal Params: {search_top.best_params_}")

# updated model
joblib.dump(best_model_top, 'optimized_frog_model_top8.pkl')

# new feature importances
importances_top = best_model_top.named_steps['xgbclassifier'].feature_importances_
pd.DataFrame({
    'Feature': top_features,
    'Importance': importances_top
}).sort_values('Importance', ascending=False).set_index('Feature').plot.barh()
plt.title('XGBoost Feature Importance (Top 8)')
plt.show()
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, classification_report, confusion_matrix

# After training the model and getting best_model_top
y_pred = best_model_top.predict(X_top)

# Calculate F1 score (both macro and weighted)
f1_macro = f1_score(y, y_pred, average='macro')
f1_weighted = f1_score(y, y_pred, average='weighted')
print(f"\nF1 Score (Macro): {f1_macro:.4f}")
print(f"F1 Score (Weighted): {f1_weighted:.4f}")

# Calculate precision and recall
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")

# Detailed classification report
print("\nDetailed Classification Report:")
print(classification_report(y, y_pred))

# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y, y_pred))

# F1 as the scoring metric for hyperparameter tuning:
from sklearn.metrics import make_scorer
f1_scorer = make_scorer(f1_score, average='weighted')

# RandomizedSearchCV with F1 scoring
search_top_f1 = RandomizedSearchCV(
    model_top, 
    param_distributions=param_grid, 
    n_iter=150,
    cv=cv,
    scoring=f1_scorer,  # Using F1 score instead of accuracy
    n_jobs=-1,
    verbose=3,
    error_score='raise'
)

print("\nRetraining model using F1 score as optimization metric...")
search_top_f1.fit(X_top, y, groups=groups)

# best model optimized for F1 score
best_model_top_f1 = search_top_f1.best_estimator_
print(f"\nBest F1 Score: {search_top_f1.best_score_:.4f}")
print(f"Optimal Parameters (F1 optimized): {search_top_f1.best_params_}")

# F1-optimized model
y_pred_f1 = best_model_top_f1.predict(X_top)
print("\nClassification Report (F1 optimized model):")
print(classification_report(y, y_pred_f1))
```
